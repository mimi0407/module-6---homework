```python
import pandas as pd
import numpy as np 
```

```python
Location = 'tamiami.csv'
df = pd.read_csv(Location)

df.head()
```

```python
columns = ['location', 'sales', 'employees', 'restaurants', 'foodcarts', 'price']

df.columns = columns
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols('price ~ sales + employees + restaurants + foodcarts', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ sales + employees + foodcarts', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ sales + employees + restaurants', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ sales + restaurants + foodcarts', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ employees + restaurants + foodcarts', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ sales + employees + restaurants + foodcarts -1', data=df).fit()
result.summary()
```

```python
result = smf.ols('price ~ restaurants + foodcarts', data=df).fit()
result.summary()
```
