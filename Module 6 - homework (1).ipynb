```python
import pandas as pd
import numpy as np
```

```python
Location = 'kc_house_data.csv'
df=pd.read_csv(Location)

df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
result = smf.ols ('grade ~ price + sqft_living + view + condition', data=df).fit()
```

```python
result.summary()
```

```python
result = smf.ols('grade ~ price + sqft_living + condition', data=df).fit()
result.summary()
```

```python
result = smf.ols('grade ~ price + view + condition', data=df).fit()
result.summary()
```

```python
result = smf.ols(formula='grade ~ price + sqft_living + view + condition - 1', data=df).fit()
result.summary()
```
